Space and time complexity measures how much space, time, speed, a program, algorthim, solution takes

Using a list(L) of size - n, and we are given a problem. 
Given a query element(q), check whether q exists in L
We compare every element of L with q, and check if there's a match
case 1: (q present in the list at index i) -------- n comparisons
cast 2: (q not present) --------- n comparisons
case 3: (q present at index n-1, last element) --------- n comparisons

In the worst case, we end up doing about n comparisons
therefore, for a list L having n elements, the worst case is n. 
similarly, best case is 1 comparison. 
But we only consider worsr case for efficiency of the code. As n increases, the number of comparisons also increase.


```python
import numpy as np
import random

l = list(range(100))
random.shuffle(l) #not sorted

l
```

```python
# search for an elemnt q in the list: O(n) where n is the length of the list
q = 31                       # 1 unit
isFound=False;               # 1 unit
for ele in l:                # n units
    if ele==31:              # 1*n units
        print("Found")       # 1*n units
        isFound=True         # 1*n units
        break;               # 1*n units
if isFound == False:         # 1 units
    print("Not Found")       # 1 unit
    
#Total time is 4n + 4. Constants dont matter as size(n) is propotional to total time. 
# therefore, time complexity is (in order of) O(n)
```

```python
#What if the list is sorted? Can we search faster?
# Show O(log n)

import math

#Source: http://www.geeksforgeeks.org/binary-search/ 
#Returns index of x in arr if present, else -1
def binarySearch (arr, l, r, x):
 
    # Check base case
    if r >= l:
 
        mid = l + math.floor((r - l)/2)
 
        # If element is present at the middle itself
        if arr[mid] == x:
            return mid
         
        # If element is smaller than mid, then it can only
        # be present in left subarray
        elif arr[mid] > x:
            return binarySearch(arr, l, mid-1, x) #the new r is the middle point
 
        # Else the element can only be present in right subarray
        else:
            return binarySearch(arr, mid+1, r, x) # the new l is the middke point
 
    else:
        # Element is not present in the array
        return -1


l.sort();
arr = l;
q =31;
binarySearch(arr,0,len(arr)-1,q)

```

```python
# Find elements common in two lists:
l1 = list(range(100))
random.shuffle(l1)


l2 = list(range(50))
random.shuffle(l2)

# find common elements : O(n*m)
cnt=0;
for i in l1:
    for j in l2:
        if i==j:
            print(i)
            cnt += 1;
print("Number of common elements:", cnt)                      
```

```python
# Find elements common in two lists:
l1 = list(range(100))
random.shuffle(l1)


l2 = list(range(50))
random.shuffle(l2)

# find common elemnts in lists in O(n) time and O(m) space if m<n

## add all elements in the smallest list into a hashtable/Dict: O(m) space
smallList = {}
for ele in l2:
    smallList[ele] = 1; # any value is OK. Key is important
    
# Now find common element 
cnt=0;
for i in l1:
    if smallList.get(i) != None: # search happens in constant time.
        print(i);
        cnt += 1;
print("Number of common elements:", cnt)                      
```
